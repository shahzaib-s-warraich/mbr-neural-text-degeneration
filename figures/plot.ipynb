```python
import matplotlib.pyplot as plt
import jsonlines
```

```python
with jsonlines.open('../results_vanilla.jsonl') as read:
    for line in read:
        print(line)
```

```python
with jsonlines.open('../results_mbr.jsonl') as read:
    for line in read:
        print(line)
```

```python
methods = ['eta', 'eps', 'top-p', 'top-k']
vanilla = [0.3574, 0.3377,0.3377, 0.3074]
mbr = [0.40366, 0.3359, 0.3359, 0.3866]

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

bars1 = axs[0].bar(methods, vanilla, color='red')
axs[0].set_title('vanilla rouge benchmark')
axs[0].set_xlabel('methods')
axs[0].set_ylabel('rouge-1')

for bar in bars1:
    height = bar.get_height()
    axs[0].annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


bars2 = axs[1].bar(methods, mbr, color='blue')
axs[1].set_title('mbr rouge benchmark')
axs[1].set_xlabel('methods')
axs[1].set_ylabel('rouge-1')

for bar in bars2:
    height = bar.get_height()
    axs[1].annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()

plt.show()
```
